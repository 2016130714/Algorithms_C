```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.linspace(1, 10000)    #start stop 사이의 숫자 50개. 등간격의 배열 생성.
plt.plot(n, n*n + 6*n + 5, label = 'n^2 + 6n + 5' )          #x축: n, y축: n^2 + 6n + 5
plt.plot(n, 1000*n, label = '1000*n')
plt.legend()
plt.show()
```

* x축이 1-10000일 경우, n^2항을 가진 그래프의 값이 훨씬 커진다. 
* n^2  vs  1000n으로, 계수 차이가 1000배인데도, n^2의 수치가 훨씬 커짐을 알 수 있다.

```python
n = np.linspace(1, 1000)
plt.plot(n, n ** 100, label = 'n^100')
plt.plot(n, 2 ** n, label = '2^n')
plt.legend()
plt.show()
```

* n^100  vs  2^n 중에서 2^n이 훨씬 크다

```python
n = np.linspace(1, 10 ** 6)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

* x축이 1부터 백만까지일 때, 루트 n이 (logn)^3보다 작게 나타난다.

```python
n = np.linspace(1, 10 ** 8)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

* 그러나 x축이 1억까지 가면 역전되어, polynomial인 n^0.5(또는 루트 n)이 커진다.
* 수치에 따라 시기가 늦을 순 있으나 결국은 polynomial이 log를 앞설 것이다.
